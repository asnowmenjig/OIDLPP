```python
#This program combines Johnson_reprint_CSVcleaning.py and Johnson_reprint_CSVcleaning_1.py and fixes an error with alpha_sect

#important stuff
#
import csv
import re
import glob
#
csv_list = glob.glob('Letter_*_counts_v5.csv')
alpha_sect = 65
#
```

```python
#defining author functions
#
print("Defining author functions...")
#
def shakespeare_confirmation(x_line):
    if re.findall('shak[^a-z]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            positive_filewriter.writerow(row)
    elif re.findall('shak[es][sp][peh]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            positive_filewriter.writerow(row)
    elif re.findall('shakof', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            positive_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            review_filewriter.writerow(row)
#
def spenser_confirmation(x_line):
    if re.findall('[\s]spens', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Spenser", x_line]
            positive_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Spenser", x_line]
            review_filewriter.writerow(row)
#
def dryden_confirmation(x_line):
    if re.findall('[^a-z]dryd', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Dryden", x_line]
            positive_filewriter.writerow(row)
    elif re.findall('ry[^d]', x_line):   #regex for obviously incorrect drydens
        with open("Dryden_duds.csv", 'a', newline = '', encoding="UTF-8") as dryden_dud_csv:
            dryden_dud_filewriter = csv.writer(dryden_dud_csv, delimiter=',')
            row = ["Dryden", x_line]
            dryden_dud_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Dryden", x_line]
            review_filewriter.writerow(row)
#
print("Success: defined author functions!")
#
```

```python
index_value = 0
#
print("Beginning data collection...")
#
for csv_list[index_value] in csv_list:
    print("Evaluating " + str(csv_list[index_value]) + "...")
    with open(csv_list[index_value], 'r') as infile_csv:
        filereader = csv.reader(infile_csv, delimiter=',')
#
        for row in filereader:
#
            if "csv_end" in row[0]:
                print("Success: done evaluating " + str(csv_list[index_value]) + "!")
#
                if alpha_sect == 73 or alpha_sect == 85:
                    alpha_sect = alpha_sect +2
                else:
                    alpha_sect = alpha_sect + 1
#
                index_value = index_value + 1
#
            else:
                if "Shakespeare" in row[0]:
                    shakespeare_confirmation(row[1])
                elif "Spenser" in row[0]:
                    spenser_confirmation(row[1])
                elif "Dryden" in row[0]:
                    dryden_confirmation(row[1])
#
print("Success: data collected!")
#
```

```python
print("Closing files..")
infile_csv.close()
print("Success: files closed!")
#
print("Congrats! The program didn't fail! Check your folder to see your results!")
#
```
