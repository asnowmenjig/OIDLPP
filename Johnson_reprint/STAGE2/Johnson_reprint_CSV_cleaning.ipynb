```python
#important stuff

import csv
import re
alpha_sect = 65
```

```python
#defining author functions

def shakespeare_confirmation(x_line):
    if re.match('shak[^a-z]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'w', newline = '', encoding="UTF-8") as positive_csv:
            filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            filewriter.writerow(row)
    elif re.match('shak[es][sp][peh]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'w', newline = '', encoding="UTF-8") as positive_csv:
            filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            filewriter.writerow(row)
    elif re.match('shakof', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'w', newline = '', encoding="UTF-8") as positive_csv:
            filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'w', newline = '', encoding="UTF-8") as review_csv:
            filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Shakespeare", x_line]
            filewriter.writerow(row)
                

                
def spenser_confirmation(x_line):
    if re.match('[\s]spens', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'w', newline = '', encoding="UTF-8") as positive_csv:
            filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Spenser", x_line]
            filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'w', newline = '', encoding="UTF-8") as review_csv:
            filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Spenser", x_line]
            filewriter.writerow(row)
                

                
                
def dryden_confirmation(x_line):
    if re.match('[^a-z]dryd', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive.csv", 'w', newline = '', encoding="UTF-8") as positive_csv:
            filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Dryden", x_line]
            filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review.csv", 'w', newline = '', encoding="UTF-8") as review_csv:
            filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Dryden", x_line]
            filewriter.writerow(row)
            
```
