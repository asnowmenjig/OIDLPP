```python
#This program combines Johnson_reprint_CSVcleaning.py and Johnson_reprint_CSVcleaning_1.py and fixes an error with alpha_sect
#Also fixes bug where csv_list was not in alphabetical order, which caused _positive.csv & _review.csv files
## to not match up to their letter equivalents
#L'Estrange, Locke, Swift, Pope

#important stuff
#
import csv
import re
import glob
#
csv_list = glob.glob('Letter_*_counts_V5.csv')
#
csv_list.sort()                                 # organizes csv_list into alphabetical order
#
print(csv_list)

alpha_sect = 65
#
```

```python
#defining author functions
#
print("Defining author functions...")
#
def lestrange_confirmation(x_line):           # [^a-z^\s]e[si\/][tir\/][r\.]
    if re.findall('l[^a-z^\s]e', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive_2.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["L\'Estrange", x_line]
            positive_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["L\'Estrange", x_line, " true"]
            review_filewriter.writerow(row)
#
def locke_confirmation(x_line):           # [ldzij\s\.1]oc[klł][eo][^a-z]
    if re.findall('[^\s][ldzij\s\.1]oc[klł][eo][^a-z]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Locke", x_line, " true"]
            positive_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_positive_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Locke", x_line]
            review_filewriter.writerow(row)
#
def swift_confirmation(x_line):           # [^a-z]sw[li][sft][tf][^a-z^\-^;]
    if re.findall('[\.][\s]sw[li][sft][tf][\.]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_positive_2.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Swift", x_line]
            positive_filewriter.writerow(row)
    elif re.findall('[a-z][\s]sw[li][sft][tf][\.]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Swift", x_line, " true"]
            review_filewriter.writerow(row)
    elif re.findall('[\s]sw[li][sft][tf][\s]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Swift", x_line, " true"]
            review_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Swift", x_line, " true"]
            review_filewriter.writerow(row)
#
def pope_confirmation(x_line):           # [^a-z]p[^i]p[ec][^a-z]
    if re.findall('pape[^a-z^\s]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as positive_csv:
            positive_filewriter = csv.writer(positive_csv, delimiter=',')
            row = ["Pope", x_line, " true"]
            positive_filewriter.writerow(row)
    elif re.findall('[fea][^a-z]p[^i]p[ec][^a-z]', x_line):
        with open("Letter_" + str(chr(alpha_sect)) + "_review_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Pope", x_line, " true"]
            review_filewriter.writerow(row)
    else:
        with open("Letter_" + str(chr(alpha_sect)) + "_positive_2.csv", 'a', newline = '', encoding="UTF-8") as review_csv:
            review_filewriter = csv.writer(review_csv, delimiter=',')
            row = ["Pope", x_line]
            review_filewriter.writerow(row)
#
print("Success: defined author functions!")
```

```python
index_value = 0
#
print("Beginning data collection...")
#
for csv_list[index_value] in csv_list:
    print("Evaluating " + str(csv_list[index_value]) + "...")
    with open(csv_list[index_value], 'r') as infile_csv:
        filereader = csv.reader(infile_csv, delimiter=',')
#
        for row in filereader:
#
            if "csv_end" in row[0]:
                print("Success: done evaluating " + str(csv_list[index_value]) + "!")
                print("Data written to Letter_" + str(chr(alpha_sect)) + "_positive.csv and Letter_" + str(chr(alpha_sect)) + "_review.csv files!")
#
                if alpha_sect == 73 or alpha_sect == 85:
                    alpha_sect = alpha_sect +2
                else:
                    alpha_sect = alpha_sect + 1
#
                index_value = index_value + 1
#
            else:
                if "L\'Estrange" in row[0]:
                    lestrange_confirmation(row[1])
                if "Locke" in row[0]:
                    locke_confirmation(row[1])
                elif "Swift" in row[0]:
                    swift_confirmation(row[1])
                elif "Pope" in row[0]:
                    pope_confirmation(row[1])
#
print("Success: data collected!")
#
```

```python
print("Closing files..")
infile_csv.close()
print("Success: files closed!")
#
print("Congrats! The program didn't fail! Check your folder to see your results!")
#
```
