```python
print("Creating Johnson_reprint_final_tally.txt...")
outfile_final_tally = open("Johnson_reprint_final_tally.txt", 'w', encoding="UTF-8")
print("Success: created Johnson_reprint_final_tally.txt!")
#
import csv
import glob
#
final_tally_list = glob.glob("Letter_*_final_tally.csv")
final_tally_list.sort()
print(final_tally_list)
#
index_value = 0
alpha_sect = 65
```

```python
#define author count variables
print("Defining author count variables...")
shakespeares = 0
spensers = 0
sidneys = 0
lestranges = 0
drydens = 0
lockes = 0
priors = 0
swifts = 0
addisons = 0
popes = 0
print("Success: defined author count variables!")
#
```

```python
print("Counting authors...")
#
for final_tally_list[index_value] in final_tally_list:
    print("Evaluating " + str(final_tally_list[index_value]) + "...")
    with open(final_tally_list[index_value], 'r') as infile_csv:
        filereader = csv.reader(infile_csv, delimiter=',')
        for row in filereader:
#
            if "csv_end" in row[0]:
#
                # define count result statements
                section_header = ("ALPHABET SECTION: " + str(chr(alpha_sect)))
                shak_counts_statement = ("Shakespeare citations = " + str(shakespeares))
                spens_counts_statement = ("Spenser citations = " + str(spensers))
                sidn_counts_statement = ("Sidney citations = " + str(sidneys))
                lestr_counts_statement = ("L\'Estrange citations " + str(lestranges))
                dryd_counts_statement = ("Dryden citations = " + str(drydens))
                locke_counts_statement = ("Locke citations = " + str(lockes))
                prior_counts_statement = ("Prior citations = " + str(priors))
                swift_counts_statement = ("Swift citations = " + str(swifts))
                addis_counts_statement = ("Addison citations = " + str(addisons))
                pope_counts_statement = ("Pope citations = " + str(popes))
                section_border = ("- - - - - - - - - - - - -")
#
                # print count results to text file
                print(section_header, file=outfile_final_tally)
                print(shak_counts_statement, file=outfile_final_tally)
                print(spens_counts_statement, file=outfile_final_tally)
                print(sidn_counts_statement, file=outfile_final_tally)
                print(lestr_counts_statement, file=outfile_final_tally)
                print(dryd_counts_statement, file=outfile_final_tally)
                print(locke_counts_statement, file=outfile_final_tally)
                print(prior_counts_statement, file=outfile_final_tally)
                print(swift_counts_statement, file=outfile_final_tally)
                print(addis_counts_statement, file=outfile_final_tally)
                print(pope_counts_statement, file=outfile_final_tally)
                print(section_border, file=outfile_final_tally)
                print("Data for alphabet section " + str(chr(alpha_sect)) + " successfully written to text file.")
#
                # redefine alpha section
                if alpha_sect == 73 or alpha_sect == 85:
                    alpha_sect = alpha_sect + 2
                else:
                    alpha_sect = alpha_sect + 1
#
                # reset citations counts
                shakespeares = 0
                spensers = 0
                sidneys = 0
                lestranges = 0
                drydens = 0
                lockes = 0
                priors = 0
                swifts = 0
                addisons = 0
                popes = 0
            
            else:
                if "Shakespeare" in row[0]:
                    shakespeares = shakespeares + 1
                elif "Spenser" in row[0]:
                    spensers = spensers + 1
                elif "Sidney" in row[0]:
                    sidneys = sidneys + 1
                elif "L\'Estrange" in row[0]:
                    lestranges = lestranges + 1
                elif "Dryden" in row[0]:
                    drydens = drydens + 1
                elif "Locke" in row[0]:
                    lockes = lockes + 1
                elif "Prior" in row[0]:
                    priors = priors + 1
                elif "Swift" in row[0]:
                    swifts = swifts + 1
                elif "Addison" in row[0]:
                    addisons = addisons + 1
                elif "Pope" in row[0]:
                    popes = popes + 1
#
print("Success: data collected!")
#
```

```python
print("Closing files..")
outfile_final_tally.close()
print("Success: files closed!")
#
print("Congrats! The program didn't fail! Check your folder to see your results!")
#
```
